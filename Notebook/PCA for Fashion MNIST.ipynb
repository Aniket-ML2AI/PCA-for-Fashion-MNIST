```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/fashion-mnist_test.csv')
```

```python
df
```

```python
X_input = df.iloc[:,1:]
```

```python
Y_target = df.iloc[:,0]
```

```python
print ('Target is',df.loc[9997,'label'])
plt.gray()
plt.matshow(np.array(df.iloc[1000,1:]).reshape(28,28))
plt.show()
```

```python
sns.heatmap(np.array(df.iloc[1000,1:]).reshape(28,28),cmap='gray_r')
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_input)
```

```python
pd.DataFrame(X_scaled)
```

```python
cov_matrix = np.cov(X_scaled.T)
```

```python
eig_value,eig_vec = np.linalg.eig(cov_matrix)
```

```python
pd.DataFrame(eig_value)
```

```python
pd.DataFrame(eig_vec)
```

```python
total_eig = sum(eig_value)
var_exp = [(i/total_eig)*100 for i in sorted(eig_value,reverse=True)]
(var_exp)
```

```python
cum_var = np.cumsum(var_exp)
pd.DataFrame(cum_var)
```

```python
from matplotlib import colors
plt.figure(figsize=(10,10))
plt.bar(range(len(var_exp)),var_exp,label='Indvidual variance explained',color='g')
plt.step(range(len(cum_var)),cum_var,label='Cumulative variance explained')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y_target, test_size=0.2, random_state=30)
```

```python
from sklearn.decomposition import PCA

pca = PCA(0.97)
X_pca = pca.fit_transform(X_scaled)
X_pca.shape
```

```python
pca.n_components_
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_pca, Y_target, test_size=0.2, random_state=30)
```

```python
from sklearn.tree import DecisionTreeClassifier
rf = DecisionTreeClassifier().fit(X_train,Y_train)
Y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_pred,Y_test)
```
